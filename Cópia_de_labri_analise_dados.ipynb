```python
import datetime


data = dataset['data']
lista_dadas = []
numero_erro = 0
format = '%d/%m/%Y'
for i in data:
    try:
      lista_dadas.append(datetime.datetime.strptime(i, format).date())
    except:
      numero_erro=numero_erro+1

df= pd.DataFrame( data=lista_dadas,columns=['Data'])
df['Data'] = pd.to_datetime(df['Data'])
df['Ano'] = df['Data'].dt.year
df['Mes'] = df['Data'].dt.month
df['Dia'] = df['Data'].dt.day
df['Frequecia_anos'] = df.groupby('Ano')['Ano'].transform('count')
#df.info()
#df
numero_erro
```

```python

import matplotlib.pyplot as fig


frequecia = df['Ano'].value_counts(sort=False).tolist()
anos = df['Ano'].astype(int).unique().tolist()
visualizar = {"Quantidade": frequecia,"anos": anos}
plt.figure(figsize=(12,9))
plt.bar(anos,frequecia)
plt.xticks(rotation=-45)
```

```python
import matplotlib.pyplot as fig
from matplotlib.colorbar import ticker


df.index = df['Frequecia_anos']
freq=fig.subplot(111)

freq.plot(df.index,df['Ano'])
fig.gcf().autofmt_xdate()
freq.xaxis.set_major_locator(ticker.MultipleLocator(500))
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

data = dataset['data']
lista_dadas= []
N_verificado = 0
Verificado = 0
format = '%d/%m/%Y'
for i in data:
    try:
      lista_dadas.append(datetime.datetime.strptime(i, format).date())
      N_verificado=N_verificado+1
    except:
      Verificado=Verificado+1

data_soma = N_verificado + Verificado
pro_NV= round(N_verificado*100/data_soma)
pro_V= round(Verificado*100/data_soma)
print(Verificado)

fig = plt.figure(
    FigureClass=Waffle, 
    rows=10, 
    values={'Datas_Não_verificado': pro_NV, 'Datas_Verificado': pro_V}, 
    colors=("#983D3D", "#232066"),
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    icons='clock', icon_size=10, 
    icon_legend=True
)

fig.set_tight_layout(False)
plt.show();
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
jr = dataset['jornal'].astype(str).unique().tolist()
  
data = dataset['jornal'].value_counts(sort=False).tolist() 
  
wp = { 'linewidth' : 0, 'edgecolor' : "green" } 
def func(pct, allvalues): 
    absolute = int(pct/100.*np.sum(allvalues)) 
    return "{:.1f}%\n({:d} g)".format(pct, absolute) 
fig, ax = plt.subplots(figsize =(15, 7)) 
wedges, texts, autotexts = ax.pie(data,  
                                  autopct = lambda pct: func(pct, data), 
                                 
                                  labels = jr, 
                                  shadow = True, 
                                  
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  
                                  textprops = dict(color ="magenta",visible ="")) 
                                  
ax.legend(wedges, jr, 
          title ="Jornais", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 5, weight ="bold") 
ax.set_title("Gráfico") 
plt.show() 
```
